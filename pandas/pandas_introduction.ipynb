### Apply

It applies function to dataframe or Series

```python
def square(x):
    return x * x
```

```python
def f(x):
    return x + 40
```

```python
def f2(x):
    return x -30
```

```python
def apply(l, f):
    for i in range(len(l)):
        e = l[i]
        l[i] = f(e)
    
    return l    
```

```python
l = [1,2,3,4]
```

```python
apply(l, square)
```

By default, the entries for which no data is available are filled with NA values. To change this, we can specify one of several options for the join and join_axes parameters of the concatenate function.
